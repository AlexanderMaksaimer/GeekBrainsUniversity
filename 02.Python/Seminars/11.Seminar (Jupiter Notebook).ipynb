Дана функция f(x)=5x^2+10x-30

```python
from sympy.solvers import solve
from sympy import Symbol
x = Symbol('x')
solve(5*x**2+10*x-30)
```

Найти интервалы, на которых функция возрастает

```python
from sympy import *
x = Symbol('x')
y = 5*x**2+10*x-30
roots = solve(5*x**2 + 10*x -30)
for root in roots:
    pprint(root,use_unicode = True)
plot(y)
```

```python
derivate = diff(y)
roots = solve(derivate)
if len(roots) ==1:
    root = roots.pop()
    if is_increasing(y,Interval(-oo,root)):
        print(f'Функция возрастает на промежутке {Interval(-oo,root)}')
    else:
        print(f'Функция убывает на промежутке {Interval(-oo,root)}')
```
